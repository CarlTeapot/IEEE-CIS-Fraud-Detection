```python
import mlflow


mlflow.set_tracking_uri("https://dagshub.com/CarlTeapot/ML_Assignment2.mlflow")


perceptron_run_id = "85ce5b8386574466ae5c016a7ab5a007"
logistic_reg_run_id = "c88776a385f94dd29c31cc0f50f9205f"
random_forest_run_id = "0e8af18d4cc840dfa1ef35d637bfbca3"
xgboost_run_id = "5b75e26c5e684a64b2ad815062f3c2fe"
hist_gradient_boosting_run_id = "7ef8a59e152e4bd8b111673579664624"
artifact_path = "best_model"

model_uri = f"runs:/{hist_gradient_boosting_run_id}/{artifact_path}"


experiments = mlflow.search_experiments()
print(f"Connected successfully to {mlflow.get_tracking_uri()}")
print(f"Found {len(experiments)} experiments")


loaded_model = mlflow.pyfunc.load_model(model_uri)
print("Model loaded successfully!")
```

```python
path = "/kaggle/input/ieee-fraud-detection"

df = pd.read_csv(path + "/test_transaction.csv")
ids = df['TransactionID']
df.drop('TransactionID', axis = 1, inplace=True)
```

```python
predictions = loaded_model.predict(df)

```

```python
submission = pd.DataFrame({
    'TransactionID': ids,
    'isFraud': predictions
})
submission.to_csv("submission.csv", index=False)
```
